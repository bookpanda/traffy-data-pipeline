```python
import pandas as pd
import requests
```

```python
url = "https://publicapi.traffy.in.th/teamchadchart-stat-api/geojson/v1"

response = requests.get(url)
response.raise_for_status()  # Raise error if request fails

geojson_data = response.json()

# Optionally print or inspect the data
print(geojson_data.keys())  # For example, print the top-level keys
```

```python
scraped_df = pd.read_csv("../data/processed_data_1k.csv")
scraped_df = scraped_df.iloc[0:0]
scraped_df
```

```python
for i in range(1000):
    sample = geojson_data["features"][i]
    ticket_id = sample["properties"]["ticket_id"]
    _type = sample["properties"]["type"]
    org = sample["properties"]["org"]
    org_action = sample["properties"]["org_action"]
    comment = sample["properties"]["description"]
    photo = sample["properties"]["photo_url"]
    # photo_after = sample["properties"]["photo_after_url"]
    address = sample["properties"]["address"]
    subdistrict = sample["properties"]["subdistrict"]
    district = sample["properties"]["district"]
    province = sample["properties"]["province"]
    timestamp = sample["properties"]["timestamp"]
    state = sample["properties"]["state"]
    star = sample["properties"]["star"]
    count_reopen = sample["properties"]["count_reopen"]
    last_activity = sample["properties"]["last_activity"]
    longitude = sample["geometry"]["coordinates"][0]
    latitude = sample["geometry"]["coordinates"][1]

    print(org, org_action)

    scraped_df.loc[len(scraped_df)] = {
        "ticket_id": ticket_id,
        "type": f"[{_type}]",
        "organization": ",".join(org),
        "comment": comment,
        "photo": photo,
        # "photo_after": photo_after,
        "address": address,
        "subdistrict": subdistrict,
        "district": district,
        "province": province,
        "timestamp": timestamp,
        "state": state,
        "star": star,
        "count_reopen": count_reopen,
        "last_activity": last_activity,
        "longitude": longitude,
        "latitude": latitude
    }
```

```python
scraped_df.to_csv("scraped_data_1k.csv", index=False)
```
