```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
engine = create_engine("postgresql://root:password@localhost:5432/pipeline_db")
query = "SELECT * FROM processed_data"

df = pd.read_sql(query, engine)
```

```python
original_df = pd.read_csv("../data/bangkok_traffy.csv")
original_df.shape
```

```python
missing_row = original_df[~original_df["ticket_id"].isin(df["ticket_id"]) & original_df["ticket_id"].notna()]
print(missing_row.dtypes)
missing_row
```

```python
missing_row["type"] = "[‡∏Å‡∏≤‡∏£‡πÄ‡∏î‡∏¥‡∏ô‡∏ó‡∏≤‡∏á]"
missing_row["latitude"] = 13.79393
missing_row["longitude"] = 100.00000
missing_row["timestamp"] = pd.to_datetime(missing_row["timestamp"])
missing_row.drop(columns=["coords"], inplace=True)
missing_row["last_activity"] = pd.to_datetime(missing_row["last_activity"])
print(missing_row.dtypes, missing_row.shape)
missing_row
```

```python
df = pd.concat([df, missing_row], ignore_index=True)
```
